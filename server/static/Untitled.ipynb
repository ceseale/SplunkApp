```python
import pandas as pd
import datetime
mainData = pd.read_csv('splunk.csv')
```

```python
data = (pd.concat ( [mainData[mainData.year == 2012],mainData[mainData.year == 2013] ] )) 
# data.to_csv('splunkReady.csv', index = False )
```

```python
def formatTime( time):
    if time < 10:
        return str(time)
    else:
        return time
    
def getPost(hour):
    if hour < 12:
        return 'AM'
    else:
        return 'PM'

line = { '_time': []}
for i in data.index.values:
    row = data.ix[i]
    toGO = str((formatTime(int (row.month)) )) +'/'+ str( formatTime(int(row.day)) ) +'/'+ str(int(row.year)) +' '+ str(int(row.hour)) + ':' + str(int(row.minute))  
    line['_time'].append(toGO) 
```

```python
weatherTable = {1: 'clear', 2: 'rain', 3:'hail', 4:'snow', 5:'dust/fog', 6:'crosswinds' }
def drunkard(item):
    if(item > 0):
        return 'drunk'
    else:
        return 'not drunk'

drunkTable = {1: 'drunk', 0:'not drunk'}


lightTable = {1: 'daylight', 2: 'dark not lighted', 3:'dark lighted', 6:'dark unknown lighting', 4:'dawn', 5:'dusk', 7:'other', 8:'not reported', 9:'unknown' }
```

```python
readyData.lgt_cond = readyData.lgt_cond.map(lightTable)
readyData.drunk_dr = readyData.drunk_dr.map(drunkard)
readyData.weather = readyData.weather.map(weatherTable)
```

```python
readyData.to_csv('accident_data.csv', index = False)
```
